## Fashion-MNIST


<b>Fashion-MNIST</b> is a dataset of Zalando's article images—consisting of a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes. We intend Fashion-MNIST to serve as a direct drop-in replacement for the original MNIST dataset for benchmarking machine learning algorithms. It shares the same image size and structure of training and testing splits.
The fashion training set we are doing today consists of 70,000 images divided into 60,000 training and 10,000 samples. Dataset sample consists of 28x28 grayscale image, associated with a label from 10 classes. The 10 classes are as follows:
<strong>
0. T-shirt/top <br>
1.Trousers <br>
2.Pullover <br>
3.Dress <br>
4.Coat <br>
5.Sandal <br>
6.Shirt <br>
7.Sneaker <br>
8.Bag  <br>
9.Ankle boot   <br>
    
Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255.

# Splitting the data
<strong>
TRAIN
VALIDATE
TEST

```python
x_train= training[:,1:]/255
y_train= training[:,0]

x_test= testing[:,1:]/255
y_test= testing[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_validate,y_train,y_validate = train_test_split(x_train,y_train,test_size=0.2,random_state=12345)
```

```python
x_train= x_train.reshape(x_train.shape[0], *(28,28,1))
x_test= x_test.reshape(x_test.shape[0], *(28,28,1))
x_validate= x_validate.reshape(x_validate.shape[0], *(28,28,1))
```

# Training the Model

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D,Dense,Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
```

### Adding Convolutional layer

```python
# no of kernal--> no of feature detector
# input share
cnn_model.add(Conv2D(32,3,3, input_shape= (28,28,1), activation='relu'))
```

### Adding Maxpooling Layer

```python
cnn_model.add(MaxPooling2D(pool_size=(2,2)))
```

### Flatten the Model

```python
cnn_model.add(Flatten())
```

### Add dense function

```python
cnn_model.add(Dense(32, activation='relu'))
```

```python
cnn_model.add(Dense(10,activation='sigmoid'))
```

## Compile

```python
cnn_model.compile(loss='sparse_categorical_crossentropy',optimizer=Adam(lr=0.001),
                 metrics=['accuracy'])
```

## Fitting the model

```python
cnn_model.fit(x_train,y_train,batch_size=512,epochs=50,verbose=1,
              validation_data=(x_validate,y_validate))
```

# Model Evaluation

```python
evaluation= cnn_model.evaluate(x_test,y_test)
print('Test Accuracy'.format(evaluation[1]))
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```
