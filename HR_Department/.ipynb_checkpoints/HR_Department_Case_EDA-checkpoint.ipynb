# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Human_Resources.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for Dupliactes
# Checking for duplicate data
duplicate_rows_df = df[df.duplicated()]
```

```python
duplicate_rows_df.shape
```

```python
# Dropping the Over18 column as that gives no more information

df=df.drop(columns='Over18')
```

```python
# Making seperate list for categorical variables
categorical_col = []
for column in df.columns:
    if df[column].dtype == object:
        categorical_col.append(column)
        print(f"{column} : {df[column].unique()}")
        print("====================================")
```

```python
df['Attrition'] = df.Attrition.astype("category").cat.codes
```

```python
df.Attrition.value_counts()
```

```python
df_numerical=df.drop(columns=['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'])
```

```python
df_numerical.columns
```

```python
df_numerical.head()
```

### Checking the effect of categorical values on the Attrition rate

```python
categorical_col
```

```python
def corr_2_cols(Col1, Col2):
    res = df.groupby([Col1, Col2]).size().unstack()
    res['perc'] = (res[res.columns[1]]/(res[res.columns[0]] + res[res.columns[1]]))
    return res
```

```python
corr_2_cols("BusinessTravel","Attrition")
```

```python
corr_2_cols("Gender","Attrition")
```

```python
corr_2_cols("Department","Attrition")
```

```python
corr_2_cols("EducationField","Attrition")
```

```python
corr_2_cols("MaritalStatus","Attrition")
```

```python
corr_2_cols("JobRole","Attrition")
```

# Visual Bi-variate Analysis.

```python
# Box plot
fig, axes = plt.subplots(4, 7, figsize = (20,15))
axes = axes.flatten()

for i in range(0,len(df_numerical.columns)):
    sns.boxplot(x="Attrition", y=df_numerical.iloc[:,i], data=df_numerical, orient='v', ax=axes[i])

plt.tight_layout()
plt.show()
```

```python
# Plotting how every feature correlate with the "target"
sns.set(font_scale=1.2)
plt.figure(figsize=(30, 30))

for i, column in enumerate(categorical_col, 1):
    plt.subplot(3, 3, i)
    g = sns.barplot(x=f"{column}", y='Attrition', data=df)
    g.set_xticklabels(g.get_xticklabels(), rotation=90)
    plt.ylabel('Attrition Count')
    plt.xlabel(f'{column}')
```

# Correlation Plot

## Checking For Multi Collinearity

```python
plt.figure(figsize=(30, 30))
sns.heatmap(df.corr(), annot=True, cmap="RdYlGn", annot_kws={"size":15})
```

### Checking for Correlation with Target Variable

```python
# Compute the correlation matrix with TARGET?RESPONSE VARIABLE
corr = df.corrwith(df.Attrition)
```

```python
corr.plot.bar(figsize=(22,8),
             title='Correlation with response variable',
             fontsize=15,rot=45, grid=True)
```
