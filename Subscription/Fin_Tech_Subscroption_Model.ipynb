# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from dateutil import parser # for columns like 'first_open' & 'enrolled_date'
```

# Importing the dataset that was created while Feature Engineering

```python
df= pd.read_csv('new_appdata10.csv')
```

```python
df.head(10)
```

### Converting the hour column from float if not scaling will not be done

```python
# First set of Feature cleaning
df["hour"] = df.hour.str.slice(1, 3).astype('int64')
```

```python
df["hour"].dtypes
```

# Data Pre-Processing

```python
x = df.loc[:, df.columns != "enrolled"]
```

```python
y = df['enrolled']
```

```python
x.head()
```

```python
y.head()
```

# Split the Dataset

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Removing the User Id column as it is not a real feature.
# but at the end we will use this data associating the prediction from user it came from

train_identifier = x_train['user']
x_train = x_train.drop(columns=['user'])

test_identifier = x_test['user']
x_test = x_test.drop(columns=['user'])
```

```python
# the user column is removed from x_train and x_test
x_train.head()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
# train the x_train and create a new variable for scaled x_train
#Standard scaler returns a numpy array so we need to convert to DataFrame
#In array form 'index' and 'column names' are lost we need it so convert to Df

x_train_scaled= pd.DataFrame(sc.fit_transform(x_train))
x_test_scaled = pd.DataFrame(sc.transform(x_test))
```

```python
# put x_train_scaled have the columns of original x_train set
x_train_scaled.columns = x_train.columns.values

# put x_test_scaled have the columns of original x_test set
x_test_scaled.columns = x_test.columns.values

# take the indexes also
x_train_scaled.index = x_train.index.values
x_test_scaled.index = x_test.index.values
```

```python
# Compare the original training set with new scaled training set
x_train= x_train_scaled
x_test= x_test_scaled
```

# Model Building

<b> Note:If working for a mobile app model always take into consideration that the screens can be correlated.
    We have tried to avoid the correlation by creating funnels but tehre may be some screens that do not belong to the same set but still be correlated.
    So add a penalty of 'L1'

```python
# Fitting Model to the Training Set
# Importing the Classifier
# Penalty to avoid correlations if some screens belong to same set.

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(penalty='l2', random_state=0)
classifier.fit(x_train, y_train)
```

```python
# Predicting Test Set
y_pred = classifier.predict(x_test)
```

# Evaluating Results


```python
# Evaluating Results
from sklearn.metrics import confusion_matrix, accuracy_score 
f1_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

# K-Fold Cross Validation 

<b>
Doing this to check if there is no Overfitting.
This is an extra step to validate the result

```python
from sklearn.model_selection import cross_val_score
accuracies= cross_val_score(estimator=classifier,X=x_train,y=y_train,cv=10)
```

```python
print('Logistic Accuracy: %0.3f (+/-%0.3f)' % (accuracies.mean(),accuracies.std()*2 ))
```

<b>
    this indicates accuracy of 76% and std devialtion of 0.009 which is very small.
    
    thus model is Accurate

```python
final_results = pd.concat([y_test, test_identifier], axis = 1).dropna()
final_results['predicted_results'] = y_pred
final_results = final_results[['user', 'enrolled', 'predicted_results']].reset_index(drop=True)
```
